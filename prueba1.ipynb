```python
import plotly.graph_objects as go
import pandas as pd
from itertools import cycle
```

```python
df = pd.DataFrame({'a': [1,2,3,4,5,6,7,8,9], 'B': [1,2,3,4,5,6,7,8,9]})
df
```

```python
def repite_trim(df):
    p = cycle([1,2,3,4])
    l = []
    for i, v in enumerate(p):
        l.append(v)
        if i >= (len(df)-1): break
        
    return l

df['Trimestre'] = repite_trim(df)
```
