```python
import pandas as pd
import datetime
import plotly.graph_objects as go
import plotly
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
import math
from itertools import cycle
import plotly.io as pio
```

```python
def repite_años(df):
    max_length =  math.ceil((len(df)+1)/4)
    length_out = len(df)
    Año=[]
    for i in range(1986, 1986 + max_length):
        Año.append(i)
        # print(i)
    Año = np.repeat(Año, 4)[:length_out]
    # print(Año)
    return Año


def repite_trimestres(df):
    l = []
    p = cycle([1, 2, 3, 4])
    for i, v in enumerate(p):
        # print(v)
        l.append(v)
        if i >= (len(df)-1):break
    return l
```

```python
link = "https://www.indec.gob.ar/ftp/cuadros/economia/sh_indicesexpgr_04.xls" #Grandes rubros trimestrales
header_gr = ["año", "ng_v", "ng_q", "ng_p", "d_1", "pp_v", "pp_q", "pp_p", "d_2","moa_v", "moa_q", "moa_p",
          "d_3", "moi_v", "moi_q", "moi_p", "d_4", "cye_v", "cye_q", "cye_p"]
GR_trimestr = pd.read_excel("./Inputs/sh_indicesexpgr_04.xls", skiprows=4, skipfooter=6, names=header_gr)
GR_trimestr["año"] =repite_años(GR_trimestr)
GR_trimestr["trimestres"] = repite_trimestres(GR_trimestr)
GR_trimestr = GR_trimestr.dropna(how='all', axis = 1)
```
